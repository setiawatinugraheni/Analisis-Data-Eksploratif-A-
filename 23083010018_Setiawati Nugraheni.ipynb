```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

file_path = "C:/Users/hani/Downloads/Dataset Pelatihan.xlsx"
data_excel = pd.read_excel(file_path)
print(data_excel.head())

nilai_data = data_excel['Nilai']

plt.figure(figsize=(10, 6))
sns.histplot(nilai_data, kde=True, bins=30, color='blue')

plt.title('Histogram dan KDE dari Kolom Nilai')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')

plt.show()
```

#### Penjelasan

berdasarkan analisis visualisasi data melalui histogram dan KDE, distribusi nilai dapat dikategorikan sebagai distribusi miring ke kri (negatively skewed). 
Karena mayoritas data berada di nilai yang lebih tinggi, yaitu antara 80 hingga 90, dengan puncak distribusi disekitar 80-85.
Ekor distribusi lebih panjang disebelah kiri, yang berarti terdapat beberapa nilai rendah (sekitar 60-70) yang menyebabkan kemiringan distribusi ke kiri.
Distribusi ini tidak simetris, dimana banyak nilai yang lebih besar dari rata-rata, tetapi sejumlah kecil data dengan nilai rendah menarik distribusi ke arah kiri.

Distribusi ini buka distribusi normal karena tidak ada keseimbangan simetris di kedua sisi puncak, dan skewness negatif menunjukkan adanya pengaruh dari data dengan nilai kecil. 
Secara keseluruhan, distribusi ini memperlihatkan pola yang menunjukkan dominasi nilai tinggi, tetapi ekor panjang di bagian nilai rendah memberikan ciri khas dari distribusi miring ke kri (negatively skewed)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

file_path = "C:/Users/hani/Downloads/Dataset Pelatihan.xlsx"
data_excel = pd.read_excel(file_path)

nilai_data = data_excel['Nilai']

mean_nilai = np.mean(nilai_data)
median_nilai = np.median(nilai_data)
modus_nilai = nilai_data.mode()[0]  # Mengambil nilai modus pertama jika ada lebih dari satu

range_nilai = np.max(nilai_data) - np.min(nilai_data)
variance_nilai = np.var(nilai_data)
std_dev_nilai = np.std(nilai_data)

skewness_nilai = skew(nilai_data)

print("=== Analisis Pemusatan Data ===")
print(f"Mean (Rata-rata): {mean_nilai}")
print(f"Median: {median_nilai}")
print(f"Modus: {modus_nilai}")

print("\n=== Analisis Penyebaran Data ===")
print(f"Range: {range_nilai}")
print(f"Variance: {variance_nilai}")
print(f"Standard Deviation: {std_dev_nilai}")

print("\n=== Skewness (Kemiringan Data) ===")
print(f"Skewness: {skewness_nilai}")

if skewness_nilai > 0:
    print("Data miring ke kanan (Positif).")
elif skewness_nilai < 0:
    print("Data miring ke kiri (Negatif).")
else:
    print("Data simetris.")
```
